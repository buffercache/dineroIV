```python
%matplotlib inline
from pandas import read_csv
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
f = open("sandy-2.csv",'r')
```

```python
df = read_csv(f)
```

```python
import seaborn as sns
sns.set_context("paper")
plt.figure(figsize=(16, 12))
```

```python
g = sns.pointplot("name", "miss_ratio", "policy", data=df[df["level"] == 2])
```

```python
g = sns.barplot("name", "miss_ratio", "policy", data=df[df["level"] == 2])
```

```python
g = sns.factorplot("name", "miss_ratio", "policy", dodge=.2, linestyles=["none", "none", "none", "none"], data=df[df["level"] == 2], size=8)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
```
